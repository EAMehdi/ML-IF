Nous n'avons pas de valeurs manquantes, mais assez peu d'observations. Commençons par travailler la données du temps. Nous nous proposons de traduire la date en nombre de jours depuis le jour 0.

```python
month_length = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
df["Date"] = pd.to_datetime(df["Date"], format="%Y-%m-%d")
df["Time"] = df["Date"].apply(lambda date: date.year*365 + (0 if date.month==1 else np.cumsum(month_length[:(date.month-1)])[-1]) + date.day)
```

Regardons chaque colonne séparéments.

```python
def explore_column(df, column, alpha=0.8, figsize=(15, 8)):
    
    def plot_histogram():
        plt.hist(df[column])
        plt.xlabel(column.capitalize())
        plt.title("Distribution de la variable %s" % column)
    
    def plot_price(alpha=alpha):
        plt.scatter(x=df[column], y=df["VKOSPI"], alpha=alpha)
        plt.xlabel(column.capitalize())
        plt.ylabel("VKOSPI")
        plt.title("VKOSPI in function of %s" % column)
    
    plt.figure(figsize=figsize)
    plt.subplot(1, 2, 1)
    plot_histogram()
    plt.subplot(1, 2, 2)
    plot_price()
    
    plt.suptitle("Variable %s" % column)
    plt.show()
    

explore_column(df, "KOSPI200")
```

Le KOSPI200 est le sous-jacent du VKOSPI et semble être inversement proportionnel à la valeur du VOKSPI. Nous avons essayé de le reproduire dans l'entraînement des données, mais ça n'a rien apporté.

Visualisons les variations dans le temps cette fois.

```python
def variation_through_time(df, column, figsize=(15, 8)):
    plt.figure(figsize=figsize)
    
    ax = plt.subplot()
    ax.plot(df["Date"], df["VKOSPI"], c=sns.color_palette()[0])
    ax.set_ylabel("VKOSPI", color=sns.color_palette()[0])
    
    ax2 = ax.twinx()
    ax2.plot(df["Date"], df[column], c=sns.color_palette()[1])
    ax2.set_ylabel("VKOSPI", color=sns.color_palette()[1])
    
    plt.show()


variation_through_time(df, "KOSPI200")
```

Il ne semble pas y avoir de lien très clair entre les deux variables, mais de large variations du KOSPI correspondent à des hausses du VKOPSI : c'est cohérent avec la définition.

## Features engineering

Pour être capable de tenir compte de ces variations, nous devons créer des features *lag* :

```python
def get_lag(df, column, lag):
    column_name = "%s - %s" % (column, lag)
    df[column_name] = np.concatenate(([np.nan]*lag, df[column][:-lag]))
    return df
```

Comme annoncé plus tôt, nous allons travailler la donnée du temps. Nous choisissons de le faire avec des sinus et des cosinus. 

De cette manière, nous serons capable de faire comprendre qu'un début de mois et proche d'une fin de mois par exemple.

```python
def get_sinus_cosinus_period(df, period, period_name):
    for operation, function in zip(["sin", "cos"], [np.sin, np.cos]):
        column_name = "%s(%s)" % (operation, period_name)
        df[column_name] = function(2 * np.pi * df["Time"] / period)
    return df
```

Après la définition de ces deux fonctions, utilisons-les :

```python
df = get_sinus_cosinus_period(df, 365, "Year")
df = get_sinus_cosinus_period(df, 30, "Month")

df = get_lag(df, "VKOSPI", 1)
df = get_lag(df, "VKOSPI", 2)
df = get_lag(df, "VKOSPI", 3)

df = get_lag(df, "KOSPI200", 1)
df = get_lag(df, "KOSPI200", 2)
df = get_lag(df, "KOSPI200", 3)

df = df.loc[3:]
```

Cela termine un bref feature engineering. Il n'avait pas pour vocation d'être exhaustif, mais donner quelques idées.

# Modélisation

Nous nous pouvons pas utiliser de cross-validation classique. Nous devons donc couper le dataset en deux à partir d'une date.

```python
df = df.sort_values(by="Date")
df = pd.get_dummies(df, columns=["Day_of_a_week"])
X = df.drop(columns=["VKOSPI", "Date"], axis=1)
y = df["VKOSPI"]



def make_historical_split(X, y, train_ratio=0.8):
    max_index = int(train_ratio * X.shape[0])
    return X.iloc[:max_index, ], X.iloc[max_index:, ], y[:max_index, ], y[max_index:, ]
```

Avec cette fonction, nous pourrons varier le ratio entre le train et le test qui nous permettra de mesurer un peu la robustesse de l'apprentissage.

Automatisons celà :

```python
from sklearn.metrics import make_scorer, mean_squared_error, r2_score



RMSE = lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true=y_true, y_pred=y_pred))

def train_predict(model, X, y, name, train_ratio=[0.5, 0.6, 0.7, 0.8, 0.9], metric=mean_squared_error):
    scores = []
    for ratio in train_ratio:
        X_train, X_test, y_train, y_test = make_historical_split(X, y, train_ratio=ratio)
        model_fitted = model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        scores.append(metric(y_true=y_test, y_pred=y_pred))
    
    mean_scores = np.mean(scores)
    std_scores = np.std(scores)
    print("%s : %0.2f (+/- %0.2f)" % (name, mean_scores, std_scores))
    return scores
```

Testons avec plusieurs algorithmes classiques :

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import ExtraTreesRegressor
from xgboost import XGBRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler


models = [
    {'model': Pipeline([('scaler', StandardScaler()), ('regression', LinearRegression())]), 'name': "Linear"},
    {'model': DecisionTreeRegressor(), 'name': "Tree"},
    {'model': DecisionTreeRegressor(max_depth=5), 'name': "Tree 5"},
    {'model': DecisionTreeRegressor(max_depth=7), 'name': "Tree 7"},
    {'model': DecisionTreeRegressor(max_depth=7), 'name': "Tree 9"},
    {'model': RandomForestRegressor(), 'name': "RF"},
    {'model': RandomForestRegressor(n_estimators=10, max_depth=5), 'name': "RF 10T-5"},
    {'model': RandomForestRegressor(n_estimators=20, max_depth=5), 'name': "RF 20T-5"},
    {'model': RandomForestRegressor(n_estimators=20, max_depth=7), 'name': "RF 20T-7"},
    {'model': ExtraTreesRegressor(n_estimators=10, max_depth=5), 'name': "ExT 10T-5"},
    {'model': ExtraTreesRegressor(n_estimators=20, max_depth=5), 'name': "ExT 20T-5"},
    {'model': ExtraTreesRegressor(n_estimators=20, max_depth=7), 'name': "ExT 20T-7"},
    {'model': XGBRegressor(n_estimators=10, max_depth=5, learning_rate=0.7, objective="reg:squarederror"), 'name': "GB 10T-5-0.7"},
    {'model': XGBRegressor(n_estimators=20, max_depth=5, learning_rate=0.7, objective="reg:squarederror"), 'name': "GB 20T-5-0.7"},
    {'model': XGBRegressor(n_estimators=20, max_depth=7, learning_rate=0.7, objective="reg:squarederror"), 'name': "GB 20T-7-0.7"},
]


for model_dict in models:
    model, name = model_dict["model"], model_dict["name"]
    train_predict(model, X, y, name, metric=RMSE)
```

Pour la RMSE , il semblerait que les Extra-Trees sont les plus performants à tous les niveaux : en RMSE et en variance.

Essayons de visualiser les prédictions dans le temps faites par les algorithmes.

```python
def visualize_predictions(model, X, y, train_ratio=0.8):
    X_train, X_test, y_train, y_test = make_historical_split(X, y, train_ratio=train_ratio)
    model = model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    max_index = int(train_ratio * X.shape[0])
    date = df["Date"][max_index:]
    
    plt.figure(figsize=(12, 8))
    plt.plot(date, y_test, label="Test")
    plt.plot(date, y_pred, label="Prediction")
    
    plt.xlabel("Temps")
    plt.ylabel("VKOSPI")
    plt.legend()
    plt.show()


model = ExtraTreesRegressor(n_estimators=20, max_depth=7)
visualize_predictions(model, X, y, train_ratio=0.9)
```

Les prédictions sont globalement proche mais semble *en retard*. Il semblerait donc que les prédictions suivent essentiellement la dernière valeur du VKOSPI. Vérifions.

Affichons l'importance de chacune des variables :

```python
def plot_feature_importance(model):
    importances = model.feature_importances_
    feature_importances = pd.Series(importances, index=X_train.columns).sort_values(ascending=False)
    
    plt.figure(figsize=(15, 8))
    feature_importances.plot.bar()
    plt.title("Feature importances using MDI")
    plt.ylabel("Mean decrease in impurity")
    plt.show()



plot_feature_importance(model)
```

Essayons avant de terminer un ensemble :

```python
from sklearn.ensemble import VotingRegressor


model = VotingRegressor([
    ('LinearRegression', Pipeline([('scaler', StandardScaler()), ('regression', LinearRegression())])), 
    ('ExtraTrees', ExtraTreesRegressor(n_estimators=20, max_depth=7)), 
    ('ExtraTrees2', ExtraTreesRegressor(n_estimators=40, max_depth=7)), 
])

scores = train_predict(model, X, y, "Voting", metric=RMSE)
```

Nous sommes meilleurs en performance mais moins bon en variance par rapport au meilleurs algorithmes réalisé.
