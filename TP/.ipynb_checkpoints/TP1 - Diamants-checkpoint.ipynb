# TP1 - Régression Linéaire : prix d'un diamant


Dataset (source : https://www.kaggle.com/datasets/shivam2503/diamonds):

* **price** : price in US dollars (\\$326-$18,823)
* **carat** : weight of the diamond (0.2-5.01)
* **cut** : quality of the cut (Fair, Good, Very Good, Premium, Ideal)
* **color** : diamond colour, from J (worst) to D (best)
* **clarity** : a measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))
* **x** : length in mm (0-10.74)
* **y** : width in mm (0-58.9)
* **z** : depth in mm (0-31.8)
* **depth** : total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43-79)
* **table** : width of top of diamond relative to widest point (43-95)

L'objectif est de prédire le prix d'un diamant à partir des informations que l'on a. Ce notebook n'a aucune prétention d'étude exhaustive du dataset mais qu'un intérêt pédagogique qui servira de support à la discussion en cours.

Pour le comprendre, nous nous proposons de constuire une fonction qui va visualiser :
1. La distribution d'une variable
2. La distribution de la variable d'intérêt *price* par rapport à la variable que l'on étudie.

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(style="whitegrid")



def explore_column(df, column, alpha=0.8, figsize=(15, 8)):
    
    def plot_histogram():
        plt.hist(df[column])
        plt.xlabel(column.capitalize())
        plt.title("Distribution de la variable %s" % column)
    
    def plot_price(alpha=alpha):
        plt.scatter(x=df[column], y=df["price"], alpha=alpha)
        plt.xlabel(column.capitalize())
        plt.ylabel("Price")
        plt.title("Price in function of %s" % column)
    
    
    
    plt.figure(figsize=figsize)
    plt.subplot(1, 2, 1)
    plot_histogram()
    plt.subplot(1, 2, 2)
    plot_price()
    
    plt.suptitle("Variable %s" % column)
    plt.show()
```

```python
explore_column(df, "y")
```

On voit clairement que nous avons deux observations qui sont très éloignées du reste des observations. De plus, ces observations ne sont pas dans la *tendance* que l'on observe entre la variable *y* et la variable *price*. Nous devons donc supprimer ces deux observations car elles vont empêcher le modèle de bien apprendre.

```python
print("Taille avant : %d" % df.shape[0])
df = df.loc[df["y"] < 30, ]
print("Taille après : %d" % df.shape[0])
```

Visualisons à présent la même chose :

## Modélisation

Après avoir mieux compris les données et savoir exploiter ces informations, il est temps de le mettre à profit. Voici comment nous allons procéder, mesurer les performances pour un modèle linéaire :
1. Qui ne considère que les données numériques
2. Qui considère les données numériques et catégorielles
3. Qui considère les features que l'on va créer suite à notre analyse des variables

Nous aurons alors trois baselines de performances. Nous essayerons d'améliorer cela en faisant une sélection d'informations, puis nous essayerons d'améliorer ces performances en regardant d'autres modèles.

Puisque nous allons tester de nombreux modèles, nous allons créer une fonction qui nous permettra de le faire en une seule ligne.

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import make_scorer, mean_squared_error, r2_score


def train_predict(model, X, y, name, metric=mean_squared_error, cv=5):
    scores = cross_val_score(model, X, y, scoring=make_scorer(metric), cv=cv)
    mean_scores = np.mean(scores)
    std_scores = np.std(scores)
    print("%s : %0.2f (+/- %0.2f)" % (name, mean_scores, std_scores))
    return mean_scores, std_scores
```

Nous allons mesurer les performances avec la métrique RMSE, il nous faut la définir. Puis nous pourrons faire nos baselines.

```python
from sklearn.linear_model import LinearRegression

RMSE = lambda y_true, y_pred: np.sqrt(mean_squared_error(y_true=y_true, y_pred=y_pred))

X = df[["carat", "depth", "table", "x", "y", "z"]]
y = df["price"]
train_predict(LinearRegression(), X, y, "Régression Linéaire - variables numériques", metric=RMSE, cv=10)

df_with_dummies = pd.get_dummies(df, columns=["cut", "color", "clarity"])
X = df_with_dummies.drop(columns=["price"], axis=1)
train_predict(LinearRegression(), X, y, "Régression Linéaire - toutes variables", metric=RMSE, cv=10)

X["x^2"] = X["x"]**2
X["y^2"] = X["y"]**2
X["z^2"] = X["z"]**2
X["carat^2"] = X["carat"]**2
train_predict(LinearRegression(), X, y, "Régression Linéaire - après étude", metric=RMSE, cv=10)
```

Essayons de faire une sélection de variables. Nous avons des idées de quoi faire avec l'études de données que nous avons fait, testons !
De même, écrivons une fonction qui nous permettra de faire plus facilement le travail.

```python
def remove_train_predict(model, X, y, columns_to_drop, name, **kwargs):
    new_X = X.drop(columns=columns_to_drop, axis=1)
    mean_scores, std_scores = train_predict(model, new_X, y, name, **kwargs)
    return mean_scores, std_scores
```

Essayons de mesurer l'apport d'une variable pour le modèle:

```python
for column in X.columns:
    name = "Régression Linéaire sans %s" % column
    remove_train_predict(LinearRegression(), X, y, [column], name, metric=RMSE, cv=10)
```

Cela semble être vraiment mieux sans *carat^2*, enlevons-là et recommençons.

```python
X = X.drop(columns=["carat^2"], axis=1)
for column in X.columns:
    name = "Régression Linéaire sans %s" % column
    remove_train_predict(LinearRegression(), X, y, [column], name, metric=RMSE, cv=10)
```

Sans *carat* et *z^2* cela semble être mieux. Recommençons :

```python
X = X.drop(columns=["carat", "z^2"], axis=1)
for column in X.columns:
    name = "Régression Linéaire sans %s" % column
    remove_train_predict(LinearRegression(), X, y, [column], name, metric=RMSE, cv=10)
```

Supprimons *depth*, et il semblerait que nous soyons bon !

```python
X = X.drop(columns=["depth"], axis=1)
for column in X.columns:
    name = "Régression Linéaire sans %s" % column
    remove_train_predict(LinearRegression(), X, y, [column], name, metric=RMSE, cv=10)
```

La nouvelle meilleure performance que l'on obtient :

```python
train_predict(LinearRegression(), X, y, "Régression Linéaire - après sélection", metric=RMSE, cv=10)
```

Nous avons donc bien amélioré sensiblement la performance ! Essayons de voir si en ajoutant de la pénalisation nous allons avoir une meilleure performance.

```python
from sklearn.linear_model import Ridge

alpha_grid = np.linspace(0.001, 2, 40)
score_result = []
for alpha in alpha_grid:
    model = Ridge(alpha=alpha)
    scores = np.sqrt(cross_val_score(model, X, y, scoring=make_scorer(mean_squared_error), cv=10))
    score_result.append(np.mean(scores))

plt.figure(figsize=(15, 8))
plt.plot(alpha_grid, score_result, 'o-')
plt.xlabel(r"Pénalisation $\lambda$ pour la régression Ridge")
plt.ylabel("Performance (RMSE)")
plt.title(r"Performance de la régression Ridge en fonction de la pénalisation $L_2$")
plt.show()
```

On voit donc bien que l'on peut améliorer la performance, mais c'est très léger par rapport à la dernière valeur de performance.
Puisque nous n'avons pas beaucoup d'informations, l'algorithme LASSO n'est pas utile ici.
